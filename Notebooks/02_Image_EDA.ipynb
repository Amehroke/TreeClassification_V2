```python
import os
import matplotlib.pyplot as plt
import rasterio
import numpy as np
from PIL import Image
```

```python
raster_filepath = r'../stockton_image_tiles/tile_-121.35410000000002_38.015079999999934.tiff'

# Open the raster file
with rasterio.open(raster_filepath) as src:
    raster_data = src.read(3)  # Read the first band
    
    print(f"Raster width: {src.width}")
    print(f"Raster height: {src.height}")
    print(f"Raster count (number of bands): {src.count}")
    print(f"Raster crs (coordinate reference system): {src.crs}")
    print(f"Raster transform: {src.transform}")

    # Display the raster data using matplotlib
    plt.imshow(raster_data, cmap='viridis')
    plt.colorbar()
    plt.title('Raster Data')
    plt.show()
```
